$1.fillna()$

    - To replace missing values with 0:
    
        - $df.fillna(0,inplace= True)
        
$3.Forward Fill (fill) and Backward Fill (fill):$

    - Use the $ffill$ method to fill missing values with the previous non-null

    - Use the $bfill$ method to fill missing values with the next non-null           value in the column.
    - To forward fill missing values:

    - Use the $interpolate()$ method.

### $Outlier$ $Analysis$

**Steps in Outlier analysis
     
   - Step 1:Find the Q1,Q2,Q3
    
        - np.percentaile(column data,q)
        
   - Step 2: Calculate the IQR 
   
        - IQR = Q3-Q1
    
   - Step 3: Calculate lower boundary and upper boundary
   
        - lb: Q1-1.5IQR
        - Ub: Q3+1.5IQR
        
   - Step 4: Find the Outliers df
   
        - c1: column data<lb
        - c2: column data>ub
        - c: appy the main condition 
        - main data[c]
        
    
    






    

```python
###########--------Step-1------#########
Q1 = np.percentile(loan_df['LoanAmount'],25)
Q2 = np.percentile(loan_df['LoanAmount'],50)
Q3 = np.percentile(loan_df['LoanAmount'],75)

##########-------Step-2-------#########
IQR = Q3-Q1

##########-------Step-3-------#########
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR

##########------Step-4--------#########
c1 = loan_df['LoanAmount'] < lb
c2 = loan_df['LoanAmount'] > ub
con = c1|c2

Loan_outliers_df = loan_df[con]
Loan_outliers_df

##########------Step-5------#######

c1 = loan_df['LoanAmount'] > lb
c2 = loan_df['LoanAmount'] < ub
con = c1&c2

Loan_non_outliers_df = loan_df[con]
Loan_non_outliers_df
```

## How to deal outliers

**Drop the outliers**

   - we can drop the outliers if outlier percentage < 2%
    
   - But this is not recommended , we lost other columns data also
    
    
**Impute with Median values**

   - As we know that Median does not affect by outliers
    
   - So it is good prcatice we can impute outliers with Median value
    
**Cap with Q3 or Q1 value**

   - If outliers are presnt less than lower bound then fill with Q1
    
   - If outliers are more than upper bound then fill with Q3

```python
# Read the each observation from prevailing wage
# if that observation <lb or >ub : fill with median value
# else: keep as it is
```

```python
# Import the packages
# Read the data

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

path=r'C:\Users\kanch\OneDrive\Naresh IT\Data science\data files\Loan Prediction\train_ctrUa4K.csv'
loan_df=pd.read_csv(path)
loan_df.head()
```

```python
###############--------step-1-------###############
Q1 = np.percentile(loan_df['LoanAmount'],25)
Q2 = np.percentile(loan_df['LoanAmount'],50)
Q3 = np.percentile(loan_df['LoanAmount'],75)

##########-------Step-2-------#########
IQR = Q3-Q1

##########-------Step-3-------#########
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR


median = loan_df['LoanAmount'].median()

list1 = []

for i in loan_df['LoanAmount']:
    if i < lb or i > ub:
        list1.append(median)
        
    else:
        list1.append(i)
loan_df['NewLoanAmount']= list1
```
