```python
# step - 1
# Take the prevaling wage column
# z-score = x -mean/sigma
# step-2: calculate mean of prevailing wage
# step-3: calculate std of prewage
# step_4: Nr: prewage-mean
#step_5: prewage_zscore=Nr/Dr
```

```python
mean=visa_df['prevailing_wage'].mean()
std = visa_df['prevailing_wage'].std()
Nr= visa_df['prevailing_wage']-mean
visa_df['prevailing_wage_z'] = Nr/std
```

**Normalization**

$minmaxscalar$

```python
#x-x_min/(x_max-x_min)
#step-1: read the pawage column
#step-2: x= data column
x = visa_df['prevailing_wage']
x_min= visa_df['prevailing_wage'].min()
x_max = visa_df['prevailing_wage'].max()
Nr = x-x_min
Dr = x_max-x_min
visa_df['prevailing_wage_norm']= Nr/Dr
```

$MinMaxScalar$

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
visa_df['prevailing_wage_mms']= mms.fit_transform(visa_df[['prevailing_wage']])
```
